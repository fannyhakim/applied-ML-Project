```python
# data2["Country or Area"] = data2["Country or Area"].str.strip()
data2 = data2[~data2["Country or Area"].isin(Areas)]
```

```python
Countries = list(data2["Country or Area"].unique()) # countries and columns filtered from the main data set
data = data[data["Area"].isin(Countries)]
data = data.drop(columns=['Area Code (M49)', 'Item Code (CPC)', 'Element Code', 'Unit'])

years = list(range(1961, 2025))

for year in years:
    str_year = str(year)
    data = data.drop(columns="Y"+str_year+"F")
    data = data.drop(columns="Y"+str_year+"N")
```
